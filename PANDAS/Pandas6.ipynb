4. Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. 쮺u치ntos de los ataques fueron letales por sexo? Devuelve un DataFrame.

```python
# Primero creramos 3 df filtrados por el pa칤s correspondiente
```

```python
df_country = df.groupby('Country').get_group('AUSTRALIA')
df_country.head()
```

```python
df_country2 = df.groupby('Country').get_group('USA')
df_country2.head()
```

```python
df_country3 = df.groupby('Country').get_group('SOUTH AFRICA')
df_country3.head()
```

```python
#맊reamos 3 data frames a partir de los de arriba, para verlos agrupados por Fatal y Sex
```

```python
# para australia

df_australia = df_country.groupby(['Fatal (Y/N)', 'Sex '])
df_australia = df_australia.size().reset_index()
df_australia['Pais'] = 'Australia'
```

```python
df_australia.head()
```

```python
# para usa

df_usa = df_country2.groupby(['Fatal (Y/N)', 'Sex '])
df_usa = df_usa.size().reset_index()
df_usa['Pais'] = 'USA'
```

```python
df_usa.head()
```

```python
# para south africa

df_south_africa = df_country3.groupby(['Fatal (Y/N)', 'Sex '])
df_south_africa = df_south_africa.size().reset_index()
df_south_africa['Pais'] = 'South Africa'
```

```python
df_south_africa.head()
```

```python
# concatenamos los 3 df para obtener el df_final con lo que pide el enunciado
```

```python
df1 = pd.concat([df_australia, df_usa], axis= 0, join= 'outer', ignore_index= True)
df_final = pd.concat([df1, df_south_africa], axis= 0, join= 'outer', ignore_index = True)
df_final.head(14)
```
