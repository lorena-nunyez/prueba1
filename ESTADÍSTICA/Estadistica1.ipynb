EJERCICIO 2.
Extraer los siguentes estadísiticos del conjunto de datos para varias de las variables disponibles:
Media
Desviación con respecto de la media
Varianza
Desviación típica
Moda
Mediana
Media ponderada
Robustez
Coeficiente de variación 

```python
media = df["salary_in_usd"].mean()
desv_media = np.abs(df["salary_in_usd"] - media)
varianza = np.var(df["salary_in_usd"])
std = np.std(df["salary_in_usd"])
moda, frecuencia = stats.mode(df["salary_in_usd"])
mediana = np.median(df["salary_in_usd"])
w = np.ones (len(df["salary_in_usd"]), dtype = np.complex64)
avg = np.average(df["salary_in_usd"], weights= w)
#robustez 
coeficiente = np.std(df["salary_in_usd"])/np.abs(media)

print(media)
print('---')
print(desv_media)
print("---")
print(varianza)
print('---')
print(std)
print('---')
print(moda)
print("---")
print(mediana)
print('---')
print(avg)
print('---')
print(coeficiente)
print('---')
```
