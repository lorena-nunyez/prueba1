EJERCICIO 1.
Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../DATA/attacks_Pandas_4.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
df_2015 = df[df["year"]>2015]
df_2015.head(20)
```

```python
actividad_agrupado = df_2015.groupby(["activity","year"])
actividad_agrupado.size()
```

```python
df_f = df_2015.groupby(["activity","year"]).size().reset_index()
df_f.head()
```

```python
plt.figure(figsize = (20,10))

fig = sns.histplot(x = 'year', y = 0 ,data = df_f, hue = 'activity')
fig.set(xlabel = 'Year', ylabel= 'Number of attacks')
fig.set_title('Number of attacks per year and activity')

plt.grid()
plt.show()
```

En la figura anterior podemos ver, por cada año, cuantos ataques han habido por la distinta actividad. Esta última está indicada con diferentes colores explicados en la leyenda, cada color representa una actividad distitnta.
Según los datos plasmados en la gráfica, los principales ataques se han dado con actividades (de más a menos) como surf skiing, surfing, feeding sharks,stand-up paddlingboarding y wading.
Además también, vemos una clara concentración de ataques en el año 2016 y otra un poco más ligera en 2017.

Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015. 

```python
df2 = df_2015.groupby(['year','sex-'])['case-number'].count().reset_index(name='nº-attacks')
df2
```

```python
plt.figure(figsize = (20,10))

fig2 = sns.histplot(x ='year',y='nº-attacks', data = df2, hue = 'sex-')
fig2.set(xlabel = 'Year', ylabel= 'Number of attacks')
fig2.set_title('Number of attacks per year and sex')
plt.grid()
plt.show()
```

En este grafico se puede ver como los hombres han sufrido más ataques que las mujeres, en todos los años estudiados

5. Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
df_pais = df.groupby('country')['case-number'].count().reset_index(name='nº-of-attacks')
df_pais
```

```python
plt.figure(figsize = (20,10))
sns.boxplot(x = 'nº-of-attacks', data = df_pais, hue= 'country')
fig.set(xlabel = 'nº of attacks')
fig.set_title('nº of attacks per country')
plt.show()
```

Con este gráfico podemos deducir, al tener estos outliers, que hay mucha diferencia de ataques de tiburones según país, como es evidente, ya que dependiendo de la zona geográfica, hay más tiburones o menos. 
En la grafica de abajo, mostramos más en detalle los ataques por país. 
Al tener tantos outliers, resulta complicado ver la información de dentro el box, como la mediana o la posición de los datos respecto a la mediana.

```python
ax = plt.subplots(1, figsize=(20, 10))
sns.boxplot(x = df_pais['country'], y = df_pais['nº-of-attacks'])
plt.xticks(rotation = 90)
plt.show()
```

```python
np.median(df_pais['nº-of-attacks'])
```

En esta grafica de aquí podemos ver que los tres países que nos marcaban los 3 outliers más 'bestias' son USA, South Africa y Australia, con más de 200 ataques (USA tiene 700), frente a los demás países que están todos por debajo de 50, siendo algunos de ellos outliers, ya que la mediana del número de ataques se concentra en 2.
